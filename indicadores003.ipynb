```python
import yfinance as yf
import pandas as pd
import ta
import matplotlib.pyplot as plt

def descargar_datos(symbol, start_date):
    # Descargar los datos históricos del activo
    data_df = yf.download(symbol, start=start_date)
    data_df.reset_index(inplace=True)
    data_df['Date'] = pd.to_datetime(data_df['Date'])
    return data_df

def calcular_indicadores(data_df):
    # Calcular indicadores técnicos relevantes
    data_df['RSI'] = ta.momentum.RSIIndicator(data_df['Close']).rsi()
    data_df['SMA'] = ta.trend.sma_indicator(data_df['Close'], window=20, fillna=True)
    data_df['EMA'] = ta.trend.ema_indicator(data_df['Close'], window=20, fillna=True)
    return data_df

def visualizar_datos(data_df):
    # Visualizar datos históricos y indicadores técnicos
    fig, ax1 = plt.subplots(figsize=(14, 7))

    color = 'tab:red'
    ax1.set_xlabel('Fecha')
    ax1.set_ylabel('Precio', color=color)
    ax1.plot(data_df['Date'], data_df['Close'], label='Precio de Cierre', color=color)
    ax1.plot(data_df['Date'], data_df['SMA'], label='SMA (20)', color='blue')
    ax1.plot(data_df['Date'], data_df['EMA'], label='EMA (20)', color='green')
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()
    color = 'tab:blue'
    ax2.set_ylabel('Volumen', color=color)
    ax2.bar(data_df['Date'], data_df['Volume'], label='Volumen', color=color, alpha=0.3)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()
    plt.title('Análisis Técnico del Activo')
    plt.show()

def inferir_decision_comercial(data_df):
    # Inferir decisión comercial basada en los indicadores
    ultimo_precio = data_df['Close'].iloc[-1]
    ultimo_rsi = data_df['RSI'].iloc[-1]
    ultimo_sma = data_df['SMA'].iloc[-1]
    ultimo_ema = data_df['EMA'].iloc[-1]
    ultimo_volumen = data_df['Volume'].iloc[-1]

    rsi_total = data_df['RSI'].mean()
    sma_total = data_df['SMA'].mean()
    ema_total = data_df['EMA'].mean()

    print(f"RSI Total: {rsi_total:.2f}, RSI Actual: {ultimo_rsi:.2f}")
    print(f"SMA Total: {sma_total:.2f}, SMA Actual: {ultimo_sma:.2f}")
    print(f"EMA Total: {ema_total:.2f}, EMA Actual: {ultimo_ema:.2f}")

    # Implementar una estrategia de trading robusta
    if ultimo_rsi < 30 and ultimo_sma < ultimo_ema and ultimo_volumen > data_df['Volume'].mean():
        return "Es preferible comprar el activo con la idea de generar ganancias."
    elif ultimo_rsi > 70 and ultimo_sma > ultimo_ema and ultimo_volumen > data_df['Volume'].mean():
        return "Es preferible vender el activo con la idea de generar ganancias."
    else:
        return "No se identifica una clara oportunidad de compra o venta."

def imprimir_combinaciones_decisiones():
    print("A partir de estos indicadores, se pueden identificar diversas combinaciones que pueden llevar a diferentes decisiones comerciales. Aquí hay algunas posibles combinaciones:")
    print()
    print("1. RSI < 30, SMA < EMA, Volumen Alto: En esta combinación, el RSI indica que el activo está sobrevendido (es decir, se considera que el precio ha caído demasiado), mientras que la SMA está por debajo de la EMA, lo que sugiere una tendencia a la baja reciente. Además, el volumen es alto, lo que puede indicar un interés significativo en el activo. Esta combinación podría interpretarse como una señal para comprar el activo, ya que se espera que el precio se recupere después de una caída excesiva.")
    print()
    print("2. RSI > 70, SMA > EMA, Volumen Alto: En este caso, el RSI indica que el activo está sobrecomprado (es decir, se considera que el precio ha subido demasiado), y tanto la SMA como la EMA están en una tendencia alcista. Además, el volumen es alto, lo que puede indicar un interés significativo en el activo. Esta combinación podría interpretarse como una señal para vender el activo, ya que es probable que haya una corrección a la baja después de una fuerte subida.")
    print()
    print("3. Otras combinaciones: Para otras combinaciones de indicadores y volumen, se puede considerar que no hay una clara oportunidad de compra o venta.")

# Definir el símbolo del activo y la fecha de inicio
symbol = 'BTC-USD'
start_date = '2024-02-20'

# Paso 1: Descargar los datos históricos
data_df = descargar_datos(symbol, start_date)

# Paso 2: Calcular indicadores técnicos
data_df = calcular_indicadores(data_df)

# Paso 3: Visualizar los datos
#visualizar_datos(data_df)

# Paso 4: Inferir la decisión comercial
decision_comercial = inferir_decision_comercial(data_df)
print(decision_comercial)

# Imprimir combinaciones de decisiones
#imprimir_combinaciones_decisiones()


'''

ERRORES
 Se deberían utilizar indicadores específicos para el análisis de tendencias,
 como medias móviles exponenciales de diferentes longitudes, MACD, estocásticos, etc.,
  y considerar la pendiente de las líneas de tendencia y los patrones de precios.
'''
```

```python
# perfecto.py
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from ta.trend import ichimoku_conversion_line, ichimoku_base_line
from ta.others import DailyReturnIndicator
from ta.momentum import UltimateOscillator
from ta.volatility import BollingerBands

class AnalisisIchimoku:
    def __init__(self, simbolo, fecha_inicio):
        self.simbolo = simbolo
        self.fecha_inicio = fecha_inicio

    def cargar_datos(self):
        data_df = yf.download(self.simbolo, start=self.fecha_inicio)
        data_df.reset_index(inplace=True)
        data_df['Date'] = pd.to_datetime(data_df['Date'])
        return data_df

    def calcular_ichimoku(self, data_df):
        data_df['Tenkan-sen'] = ichimoku_conversion_line(data_df['High'], data_df['Low'])
        data_df['Kijun-sen'] = ichimoku_base_line(data_df['High'], data_df['Low'])
        return data_df

    def calcular_bandas_bollinger(self, data_df):
        bb = BollingerBands(close=data_df['Close'])
        data_df['BB_Middle'] = bb.bollinger_mavg()
        data_df['BB_Upper'] = bb.bollinger_hband()
        data_df['BB_Lower'] = bb.bollinger_lband()
        return data_df

    def interpretar_señales(self, data_df):
        tenkan_sen_hoy = data_df['Tenkan-sen'].iloc[-1]
        kijun_sen_hoy = data_df['Kijun-sen'].iloc[-1]

        tendencia = "neutral"
        if tenkan_sen_hoy > kijun_sen_hoy:
            tendencia = "alcista"
        elif tenkan_sen_hoy < kijun_sen_hoy:
            tendencia = "bajista"

        # Calcular la señal de compra o venta
        bb_signal = "Compra" if data_df['Close'].iloc[-1] > data_df['BB_Upper'].iloc[-1] else "Venta"

        confianza = "Desconocida"
        if bb_signal == "Compra":
            confianza = "Fuerte" if tendencia == "alcista" else "Débil"
        elif bb_signal == "Venta":
            confianza = "Fuerte" if tendencia == "bajista" else "Débil"

        recomendacion = f"{confianza} {tendencia.capitalize()}"

        return recomendacion

    def calcular_stop_take(self, data_df):
        # Calcular stop loss y take profit como un porcentaje del rango de precios reciente
        ultimo_cierre = data_df['Close'].iloc[-1]
        rango = data_df['High'].max() - data_df['Low'].min()
        # Ajustar por volatilidad del activo
        volatilidad = data_df['Close'].std()
        porcentaje_stop_loss = 0.03 * (volatilidad / 10)  # Porcentaje ajustado por volatilidad para el stop loss
        porcentaje_take_profit = 0.05 * (volatilidad / 10)  # Porcentaje ajustado por volatilidad para el take profit
        stop_loss = ultimo_cierre - porcentaje_stop_loss
        take_profit = ultimo_cierre + porcentaje_take_profit
        return stop_loss, take_profit

    def generar_grafico(self, data_df, ultimate_oscillator, bb_signal, bollinger_bands):
        fig = go.Figure()

        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Close'], mode='lines', name='Precio de Cierre', line=dict(color='black')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Tenkan-sen'], mode='lines', name='Tenkan-sen', line=dict(color='blue')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Kijun-sen'], mode='lines', name='Kijun-sen', line=dict(color='red')))

        # Calcular y graficar el indicador de retorno diario
        dr_indicator = DailyReturnIndicator(close=data_df['Close'])
        daily_return = dr_indicator.daily_return()
        fig.add_trace(go.Scatter(x=data_df['Date'], y=daily_return, mode='lines', name='Retorno Diario', line=dict(color='green')))

        # Graficar el Ultimate Oscillator
        fig.add_trace(go.Scatter(x=data_df['Date'], y=ultimate_oscillator, mode='lines', name='Ultimate Oscillator', line=dict(color='purple')))

        # Graficar las bandas de Bollinger
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Middle'], mode='lines', name='Bollinger Middle Band', line=dict(color='orange')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Upper'], mode='lines', name='Bollinger Upper Band', line=dict(color='red')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Lower'], mode='lines', name='Bollinger Lower Band', line=dict(color='blue')))

        fig.update_layout(title='Ichimoku Kinkō Hyō (Ichimoku), Retorno Diario, Ultimate Oscillator y Bollinger Bands',
                          xaxis_title='Fecha',
                          yaxis_title='Precio')

        fig.show()

    def ejecutar_analisis(self):
        data_df = self.cargar_datos()
        data_df = self.calcular_ichimoku(data_df)
        data_df = self.calcular_bandas_bollinger(data_df)

        ultimate = UltimateOscillator(high=data_df['High'], low=data_df['Low'], close=data_df['Close'])
        ultimate_oscillator = ultimate.ultimate_oscillator()

        tendencia = self.interpretar_señales(data_df)

        stop_loss, take_profit = self.calcular_stop_take(data_df)
        print("Tendencia actual:", tendencia)
        print("Recomendación:", tendencia)

        print("Stop Loss:", stop_loss)
        print("Take Profit:", take_profit)

        bollinger_bands = data_df[['BB_Middle', 'BB_Upper', 'BB_Lower']]
        #self.generar_grafico(data_df, ultimate_oscillator, None, bollinger_bands)
if __name__ == "__main__":
    simbolo = 'BTC-USD'
    fecha_inicio = '2023-04-15'

    analisis_ichimoku = AnalisisIchimoku(simbolo, fecha_inicio)
    analisis_ichimoku.ejecutar_analisis()
```

```python
# perfecto.py
import yfinance as yf
import pandas_ta as ta
import plotly.graph_objects as go
from datetime import datetime, timedelta

class EstrategiaTrading:
    def __init__(self, simbolo, fecha_inicio, fecha_fin, intervalo='1m', dias=1):
        self.simbolo = simbolo
        self.fecha_inicio = fecha_inicio
        self.fecha_fin = fecha_fin
        self.intervalo = intervalo
        self.dias = dias
        self.capital_total = 20000  # Capital total disponible para trading
        self.atr_length = None  # Longitud para el cálculo del ATR (se actualizará dinámicamente)
        self.atr_multiplier_stop_loss = 2  # Multiplicador para el cálculo del stop-loss basado en ATR
        self.atr_multiplier_take_profit = 3  # Multiplicador para el cálculo del take-profit basado en ATR
        self.porcentaje_max_posicion = 0.02  # Porcentaje máximo del capital total en una posición

    def descargar_datos(self):
        return yf.download(self.simbolo, start=self.fecha_inicio, end=self.fecha_fin, interval=self.intervalo)

    def identificar_tendencia(self, df):
        df['EMA_corta'] = ta.ema(df['Close'], length=20)
        df['EMA_larga'] = ta.ema(df['Close'], length=50)
        df['Tendencia'] = 'Indefinida'
        df.loc[df['EMA_corta'] > df['EMA_larga'], 'Tendencia'] = 'Alcista'
        df.loc[df['EMA_corta'] < df['EMA_larga'], 'Tendencia'] = 'Bajista'
        return df

    def calcular_rsi(self, df):
        df['RSI'] = ta.rsi(df['Close'], length=14)
        return df

    def evaluar_senales(self, df):
        if self.generar_senal_compra(df):
            self.gestionar_posicion(df, "comprar")
        elif self.generar_senal_venta(df):
            self.gestionar_posicion(df, "vender")
        else:
            print("No se detectaron señales de compra o venta.")

        self.detectar_evento_liquidez(df)

    def generar_senal_compra(self, df):
        return (df['EMA_corta'].iloc[-1] > df['EMA_larga'].iloc[-1]) and (df['RSI'].iloc[-1] < 30)

    def generar_senal_venta(self, df):
        return (df['EMA_corta'].iloc[-1] < df['EMA_larga'].iloc[-1]) and (df['RSI'].iloc[-1] > 70)

    def gestionar_posicion(self, df, accion):
        if self.atr_length == 0:
            print("Error: La longitud del ATR es cero. No se puede calcular el tamaño de la posición.")
            return

        riesgo_por_operacion = self.capital_total * self.porcentaje_max_posicion
        tamano_posicion = riesgo_por_operacion / self.atr_length

        if accion == "comprar":
            precio_compra = df.iloc[-1]['Close']
            stop_loss = precio_compra - (self.atr_multiplier_stop_loss * self.atr_length)
            take_profit = precio_compra + (self.atr_multiplier_take_profit * self.atr_length)
            print(f"Señal de compra detectada. Compra ahora a {precio_compra}, Stop-loss en {stop_loss}, Take-profit en {take_profit}")
            # Implementa la lógica para abrir una posición de compra

        elif accion == "vender":
            precio_venta = df.iloc[-1]['Close']
            stop_loss = precio_venta + (self.atr_multiplier_stop_loss * self.atr_length)
            take_profit = precio_venta - (self.atr_multiplier_take_profit * self.atr_length)
            print(f"Señal de venta detectada. Vende ahora a {precio_venta}, Stop-loss en {stop_loss}, Take-profit en {take_profit}")
            # Implementa la lógica para abrir una posición de venta

    def visualizar_datos(self, df):
        fig = go.Figure(data=[go.Candlestick(x=df.index,
                                            open=df['Open'],
                                            high=df['High'],
                                            low=df['Low'],
                                            close=df['Close'])])

        fig.add_trace(go.Scatter(x=df.index, y=df['EMA_corta'], mode='lines', name='EMA (Corta)', line=dict(color='blue')))
        fig.add_trace(go.Scatter(x=df.index, y=df['EMA_larga'], mode='lines', name='EMA (Larga)', line=dict(color='red')))

        niveles_fibonacci = self.calcular_niveles_fibonacci(df)
        for nivel, precio in niveles_fibonacci.items():
            fig.add_shape(
                type="line",
                x0=df.index[0],
                y0=precio,
                x1=df.index[-1],
                y1=precio,
                line=dict(color='green', width=1, dash='dash'),
                name=f'Fib {nivel}'
            )

        fig.update_layout(title='Tendencia del precio (EMA) con Retracement de Fibonacci',
                          xaxis_rangeslider_visible=False)
        fig.show()

    def estrategia_reversion_media(self, df, ventana):
        # Calcular la media móvil exponencial
        df['Media'] = ta.ema(df['Close'], length=ventana)

        # Calcular la desviación estándar
        df['Desviacion'] = ta.stdev(df['Close'], length=ventana)

        # Calcular las bandas de Bollinger con la media móvil exponencial
        df['Banda_Superior'] = df['Media'] + 2 * df['Desviacion']
        df['Banda_Inferior'] = df['Media'] - 2 * df['Desviacion']

        # Generar señales de compra y venta
        df['Señal_Compra'] = (df['Close'] < df['Banda_Inferior']).astype(int)
        df['Señal_Venta'] = (df['Close'] > df['Banda_Superior']).astype(int)

        return df

    def cruce_medias_moviles(self, df, corta, larga):
        df['corta'] = ta.ema(df['Close'], length=corta)
        df['larga'] = ta.ema(df['Close'], length=larga)
        df['Posicion'] = 0

        df.loc[df['corta'] > df['larga'], 'Posicion'] = 1
        df.loc[df['corta'] < df['larga'], 'Posicion'] = -1

        return df

    def calcular_bbands(self, df):
        # Calcular la media móvil simple de 20 periodos
        df['SMA'] = ta.sma(df['Close'], length=20)

        # Calcular la desviación estándar de 20 periodos
        df['STD'] = ta.stdev(df['Close'], length=20)

        # Calcular las bandas de Bollinger
        df['BB_UPPER'] = df['SMA'] + 2 * df['STD']
        df['BB_LOWER'] = df['SMA'] - 2 * df['STD']

        return df

    def calcular_atr(self, df):
        self.verificar_datos_completos(df)  # Verificar que los datos sean válidos
        self.atr_length = ta.atr(df['High'], df['Low'], df['Close'], length=14).iloc[-1]

    def verificar_datos_completos(self, df):
        if df[['High', 'Low', 'Close']].isnull().values.any():
            raise ValueError("Hay valores faltantes en las columnas 'High', 'Low' o 'Close' del DataFrame.")

    def detectar_evento_liquidez(self, df):
        self.calcular_atr(df)  # Calcula el ATR antes de utilizarlo

        # Comparar el rango verdadero promedio con el cierre anterior para detectar eventos de liquidez
        if df.iloc[-1]['Close'] > df.iloc[-2]['Close'] + self.atr_length or \
           df.iloc[-1]['Close'] < df.iloc[-2]['Close'] - self.atr_length:
            print("¡Se detectó un evento de liquidez!")

    def calcular_niveles_fibonacci(self, df):
        maximo = df['High'].max()
        minimo = df['Low'].min()
        niveles_fibonacci = {
            0: minimo,
            0.236: maximo - (0.236 * (maximo - minimo)),
            0.382: maximo - (0.382 * (maximo - minimo)),
            0.5: maximo - (0.5 * (maximo - minimo)),
            0.618: maximo - (0.618 * (maximo - minimo)),
            0.786: maximo - (0.786 * (maximo - minimo)),
            1: maximo
        }
        return niveles_fibonacci

    def simular_trading_retroactivo(self):
        # Descargar datos históricos
        df = self.descargar_datos()

        # Identificar tendencia
        df = self.identificar_tendencia(df)

        # Calcular indicadores EMA, BBANDS y RSI
        df = self.calcular_bbands(df)
        df = self.calcular_rsi(df)

        # Calcular ATR y detectar eventos de liquidez
        self.detectar_evento_liquidez(df)

        # Evaluar señales de compra o venta
        self.evaluar_senales(df)

        # Visualizar patrones de velas y señales con Retracement de Fibonacci
        self.visualizar_datos(df)

        # Implementar estrategia de cruce de medias móviles
        df = self.cruce_medias_moviles(df, 20, 50)

        # Implementar estrategia de reversión a la media
        df = self.estrategia_reversion_media(df, ventana=20)

# Configuración de parámetros
simbolo = "DOT-USD" #@param ["BTC-USD", "ETH-USD", "ADA-USD", "SOL-USD", "DOT-USD", "USDT-USD", "DAI-USD", "USDC-USD", "LUNA-USD", "MATIC-USD", "AVAX-USD", "BNB-USD", "MANA-USD", "SAND-USD", "SLP-USD", "AXS-USD", "UNI-USD", "ALGO-USD", "AAVE-USD", "ENS-USD", "NEAR-USD", "FTM-USD", "PAXG-USD", "DOGE-USD", "TRX-USD", "ARB-USD", "LTC-USD", "XLM-USD", "OP-USD", "TRX-USD", "LUNA-USD"]
dias = "15"  #@param ["59", "15", "7", "1"]
intervalo = "15m" #@param ["1m", "5m", "15m", "1d"]

# Calcular fechas de inicio y fin
fecha_inicio = datetime.now() - timedelta(days=int(dias))
fecha_fin = datetime.now()

# Crear instancia de la estrategia de trading
estrategia = EstrategiaTrading(simbolo, fecha_inicio, fecha_fin, intervalo, dias)

# Realizar simulación retroactiva
estrategia.simular_trading_retroactivo()
```

```python
# perfecto.py
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from ta.trend import ichimoku_conversion_line, ichimoku_base_line
from ta.others import DailyReturnIndicator
from ta.momentum import UltimateOscillator
from ta.volatility import BollingerBands

class AnalisisIchimoku:
    def __init__(self, simbolo, fecha_inicio):
        self.simbolo = simbolo
        self.fecha_inicio = fecha_inicio

    def cargar_datos(self):
        data_df = yf.download(self.simbolo, start=self.fecha_inicio)
        data_df.reset_index(inplace=True)
        return data_df

    def calcular_ichimoku(self, data_df):
        data_df['Date'] = pd.to_datetime(data_df['Date'])
        data_df['Tenkan-sen'] = ichimoku_conversion_line(data_df['High'], data_df['Low'])
        data_df['Kijun-sen'] = ichimoku_base_line(data_df['High'], data_df['Low'])
        return data_df

    def calcular_bandas_bollinger(self, data_df):
        bb = BollingerBands(close=data_df['Close'])
        data_df['BB_Middle'] = bb.bollinger_mavg()
        data_df['BB_Upper'] = bb.bollinger_hband()
        data_df['BB_Lower'] = bb.bollinger_lband()
        return data_df

    def interpretar_señales(self, data_df):
        tenkan_sen_hoy = data_df['Tenkan-sen'].iloc[-1]
        kijun_sen_hoy = data_df['Kijun-sen'].iloc[-1]

        tendencia = "neutral"
        if tenkan_sen_hoy > kijun_sen_hoy:
            tendencia = "alcista"
        elif tenkan_sen_hoy < kijun_sen_hoy:
            tendencia = "bajista"

        # Calcular la señal de compra o venta
        bb_signal = "Compra" if data_df['Close'].iloc[-1] > data_df['BB_Upper'].iloc[-1] else "Venta"

        confianza = "Desconocida"
        if bb_signal == "Compra":
            confianza = "Fuerte" if tendencia == "alcista" else "Débil"
        elif bb_signal == "Venta":
            confianza = "Fuerte" if tendencia == "bajista" else "Débil"

        recomendacion = f"{confianza} {tendencia.capitalize()}"

        return recomendacion

    def calcular_stop_take(self, data_df):
        # Calcular stop loss y take profit como un porcentaje del rango de precios reciente
        ultimo_cierre = data_df['Close'].iloc[-1]
        rango = data_df['High'].max() - data_df['Low'].min()
        # Ajustar por volatilidad del activo
        volatilidad = data_df['Close'].std()
        porcentaje_stop_loss = 0.03 * (volatilidad)  # Porcentaje ajustado por volatilidad para el stop loss
        porcentaje_take_profit = 0.05 * (volatilidad)  # Porcentaje ajustado por volatilidad para el take profit
        stop_loss = ultimo_cierre - porcentaje_stop_loss
        take_profit = ultimo_cierre + porcentaje_take_profit
        return stop_loss, take_profit

    def generar_grafico(self, data_df, ultimate_oscillator, bollinger_bands):
        fig = go.Figure()

        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Close'], mode='lines', name='Precio de Cierre', line=dict(color='black')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Tenkan-sen'], mode='lines', name='Tenkan-sen', line=dict(color='blue')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Kijun-sen'], mode='lines', name='Kijun-sen', line=dict(color='red')))

        # Calcular y graficar el indicador de retorno diario
        dr_indicator = DailyReturnIndicator(close=data_df['Close'])
        daily_return = dr_indicator.daily_return()
        fig.add_trace(go.Scatter(x=data_df['Date'], y=daily_return, mode='lines', name='Retorno Diario', line=dict(color='green')))

        # Graficar el Ultimate Oscillator
        fig.add_trace(go.Scatter(x=data_df['Date'], y=ultimate_oscillator, mode='lines', name='Ultimate Oscillator', line=dict(color='purple')))

        # Graficar las bandas de Bollinger
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Middle'], mode='lines', name='Bollinger Middle Band', line=dict(color='orange')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Upper'], mode='lines', name='Bollinger Upper Band', line=dict(color='red')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Lower'], mode='lines', name='Bollinger Lower Band', line=dict(color='blue')))

        fig.update_layout(title='Ichimoku Kinkō Hyō (Ichimoku), Retorno Diario, Ultimate Oscillator y Bollinger Bands',
                          xaxis_title='Fecha',
                          yaxis_title='Precio')

        fig.show()

    def ejecutar_analisis(self):
        data_df = self.cargar_datos()
        data_df = self.calcular_ichimoku(data_df)
        data_df = self.calcular_bandas_bollinger(data_df)

        ultimate = UltimateOscillator(high=data_df['High'], low=data_df['Low'], close=data_df['Close'])
        ultimate_oscillator = ultimate.ultimate_oscillator()

        tendencia = self.interpretar_señales(data_df)

        stop_loss, take_profit = self.calcular_stop_take(data_df)
        print("Tendencia actual:", tendencia)
        print("Recomendación:", tendencia)

        print("Stop Loss:", stop_loss)
        print("Take Profit:", take_profit)

        bollinger_bands = data_df[['BB_Middle', 'BB_Upper', 'BB_Lower']]
        self.generar_grafico(data_df, ultimate_oscillator, bollinger_bands)

        # Calcular el resultado total promedio y último de todos los indicadores
        total_resultados = data_df[['Tenkan-sen', 'Kijun-sen', 'Close', 'BB_Middle', 'BB_Upper', 'BB_Lower']].mean()
        ultimo_resultado = data_df[['Tenkan-sen', 'Kijun-sen', 'Close', 'BB_Middle', 'BB_Upper', 'BB_Lower']].iloc[-1]
        print("\nResultados Totales Promedio de Todos los Indicadores:")
        print(total_resultados)
        print("\nÚltimos Resultados de Todos los Indicadores:")
        print(ultimo_resultado)

if __name__ == "__main__":
    simbolo = 'BTC-USD'
    fecha_inicio = '2024-04-05'

    analisis_ichimoku = AnalisisIchimoku(simbolo, fecha_inicio)
    analisis_ichimoku.ejecutar_analisis()
```

```python
# perfecto.py
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from ta.trend import ichimoku_conversion_line, ichimoku_base_line
from ta.others import DailyReturnIndicator
from ta.momentum import UltimateOscillator
from ta.volatility import BollingerBands

class AnalisisIchimoku:
    def __init__(self, simbolo, fecha_inicio):
        self.simbolo = simbolo
        self.fecha_inicio = fecha_inicio

    def cargar_datos(self):
        data_df = yf.download(self.simbolo, start=self.fecha_inicio)
        data_df.reset_index(inplace=True)
        # Calcular Senkou Span A y Senkou Span B necesarios para Ichimoku
        data_df['Senkou Span A'] = (data_df['High'] + data_df['Low']) / 2  # Cambia esto según el cálculo real
        data_df['Senkou Span B'] = (data_df['High'] + data_df['Low'] + data_df['Close']) / 3  # Cambia esto según el cálculo real
        return data_df

    def calcular_ichimoku(self, data_df):
        data_df['Date'] = pd.to_datetime(data_df['Date'])
        data_df['Tenkan-sen'] = ichimoku_conversion_line(data_df['High'], data_df['Low'])
        data_df['Kijun-sen'] = ichimoku_base_line(data_df['High'], data_df['Low'])

        # Calcular niveles de soporte y resistencia de la nube de Ichimoku
        data_df['Cloud_Top'] = data_df[['Senkou Span A', 'Senkou Span B']].max(axis=1)
        data_df['Cloud_Bottom'] = data_df[['Senkou Span A', 'Senkou Span B']].min(axis=1)

        # Calcular niveles de soporte y resistencia de Tenkan-sen y Kijun-sen
        data_df['Tenkan_Support'] = data_df[['Tenkan-sen', 'Kijun-sen']].min(axis=1)
        data_df['Kijun_Resistance'] = data_df[['Tenkan-sen', 'Kijun-sen']].max(axis=1)

        return data_df

    def calcular_bandas_bollinger(self, data_df):
        bb = BollingerBands(close=data_df['Close'])
        data_df['BB_Middle'] = bb.bollinger_mavg()
        data_df['BB_Upper'] = bb.bollinger_hband()
        data_df['BB_Lower'] = bb.bollinger_lband()
        return data_df

    def interpretar_señales(self, data_df):
        tenkan_sen_hoy = data_df['Tenkan-sen'].iloc[-1]
        kijun_sen_hoy = data_df['Kijun-sen'].iloc[-1]

        # Identificar niveles de soporte y resistencia
        if data_df['Close'].iloc[-1] > data_df['Cloud_Top'].iloc[-1]:
            soporte = data_df['Cloud_Top'].iloc[-1]
            resistencia = data_df['Cloud_Bottom'].iloc[-1]
        else:
            soporte = data_df['Cloud_Bottom'].iloc[-1]
            resistencia = data_df['Cloud_Top'].iloc[-1]

        tendencia = "neutral"
        if tenkan_sen_hoy > kijun_sen_hoy:
            tendencia = "alcista"
        elif tenkan_sen_hoy < kijun_sen_hoy:
            tendencia = "bajista"

        # Calcular la señal de compra o venta
        bb_signal = "Compra" if data_df['Close'].iloc[-1] > data_df['BB_Upper'].iloc[-1] else "Venta"

        confianza = "Desconocida"
        if bb_signal == "Compra":
            confianza = "Fuerte" if tendencia == "alcista" else "Débil"
        elif bb_signal == "Venta":
            confianza = "Fuerte" if tendencia == "bajista" else "Débil"

        recomendacion = f"{confianza} {tendencia.capitalize()}"

        return soporte, resistencia, recomendacion

    def calcular_stop_take(self, data_df):
        # Calcular stop loss y take profit como un porcentaje del rango de precios reciente
        ultimo_cierre = data_df['Close'].iloc[-1]
        rango = data_df['High'].max() - data_df['Low'].min()
        # Ajustar por volatilidad del activo
        volatilidad = data_df['Close'].std()
        porcentaje_stop_loss = 0.03 * (volatilidad)  # Porcentaje ajustado por volatilidad para el stop loss
        porcentaje_take_profit = 0.05 * (volatilidad)  # Porcentaje ajustado por volatilidad para el take profit
        stop_loss = ultimo_cierre - porcentaje_stop_loss
        take_profit = ultimo_cierre + porcentaje_take_profit
        return stop_loss, take_profit

    def generar_grafico(self, data_df, ultimate_oscillator, bollinger_bands):
        fig = go.Figure()

        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Close'], mode='lines', name='Precio de Cierre', line=dict(color='black')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Tenkan-sen'], mode='lines', name='Tenkan-sen', line=dict(color='blue')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Kijun-sen'], mode='lines', name='Kijun-sen', line=dict(color='red')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Cloud_Top'], mode='lines', name='Nube - Resistencia', line=dict(color='green')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Cloud_Bottom'], mode='lines', name='Nube - Soporte', line=dict(color='orange')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Tenkan_Support'], mode='lines', name='Soporte Tenkan-sen', line=dict(color='purple')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Kijun_Resistance'], mode='lines', name='Resistencia Kijun-sen', line=dict(color='brown')))

        # Calcular y graficar el indicador de retorno diario
        dr_indicator = DailyReturnIndicator(close=data_df['Close'])
        daily_return = dr_indicator.daily_return()
        fig.add_trace(go.Scatter(x=data_df['Date'], y=daily_return, mode='lines', name='Retorno Diario', line=dict(color='green')))

        # Graficar el Ultimate Oscillator
        fig.add_trace(go.Scatter(x=data_df['Date'], y=ultimate_oscillator, mode='lines', name='Ultimate Oscillator', line=dict(color='purple')))

        # Graficar las bandas de Bollinger
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Middle'], mode='lines', name='Bollinger Middle Band', line=dict(color='orange')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Upper'], mode='lines', name='Bollinger Upper Band', line=dict(color='red')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=bollinger_bands['BB_Lower'], mode='lines', name='Bollinger Lower Band', line=dict(color='blue')))

        fig.update_layout(title='Ichimoku Kinkō Hyō (Ichimoku), Retorno Diario, Ultimate Oscillator y Bollinger Bands',
                          xaxis_title='Fecha',
                          yaxis_title='Precio')

        fig.show()

    def ejecutar_analisis(self):
        data_df = self.cargar_datos()
        data_df = self.calcular_ichimoku(data_df)
        data_df = self.calcular_bandas_bollinger(data_df)

        ultimate = UltimateOscillator(high=data_df['High'], low=data_df['Low'], close=data_df['Close'])
        ultimate_oscillator = ultimate.ultimate_oscillator()

        soporte, resistencia, tendencia = self.interpretar_señales(data_df)

        stop_loss, take_profit = self.calcular_stop_take(data_df)
        print("Tendencia actual:", tendencia)
        print("Soporte:", soporte)
        print("Resistencia:", resistencia)

        print("Stop Loss:", stop_loss)
        print("Take Profit:", take_profit)

        bollinger_bands = data_df[['BB_Middle', 'BB_Upper', 'BB_Lower']]
        self.generar_grafico(data_df, ultimate_oscillator, bollinger_bands)

        # Calcular el resultado total promedio y último de todos los indicadores
        total_resultados = data_df[['Tenkan-sen', 'Kijun-sen', 'Close', 'BB_Middle', 'BB_Upper', 'BB_Lower']].mean()
        ultimo_resultado = data_df[['Tenkan-sen', 'Kijun-sen', 'Close', 'BB_Middle', 'BB_Upper', 'BB_Lower']].iloc[-1]
        print("\nResultados Totales Promedio de Todos los Indicadores:")
        print(total_resultados)
        print("\nÚltimos Resultados de Todos los Indicadores:")
        print(ultimo_resultado)

if __name__ == "__main__":
    simbolo = 'BTC-USD'
    fecha_inicio = '2024-04-05'

    analisis_ichimoku = AnalisisIchimoku(simbolo, fecha_inicio)
    analisis_ichimoku.ejecutar_analisis()
```

```python
# perfecto.py
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from ta.trend import VortexIndicator

class AnalisisVortex:
    def __init__(self, simbolo, fecha_inicio):
        self.simbolo = simbolo
        self.fecha_inicio = fecha_inicio

    def cargar_datos(self):
        data_df = yf.download(self.simbolo, start=self.fecha_inicio)
        data_df.reset_index(inplace=True)
        return data_df

    def calcular_vortex(self, data_df):
        vi = VortexIndicator(high=data_df['High'], low=data_df['Low'], close=data_df['Close'])
        data_df['Vortex_Positive'] = vi.vortex_indicator_pos()
        data_df['Vortex_Negative'] = vi.vortex_indicator_neg()
        return data_df

    def interpretar_resultados(self, data_df):
        ultimo_vortex_pos = data_df['Vortex_Positive'].iloc[-1]
        ultimo_vortex_neg = data_df['Vortex_Negative'].iloc[-1]

        mensaje = "La tendencia es neutral."
        if ultimo_vortex_pos > ultimo_vortex_neg:
            mensaje = "La tendencia es alcista."
        elif ultimo_vortex_pos < ultimo_vortex_neg:
            mensaje = "La tendencia es bajista."

        return mensaje

    def generar_grafico(self, data_df):
        fig = go.Figure()

        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Close'], mode='lines', name='Precio de Cierre', line=dict(color='black')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Vortex_Positive'], mode='lines', name='Vortex Positivo', line=dict(color='blue')))
        fig.add_trace(go.Scatter(x=data_df['Date'], y=data_df['Vortex_Negative'], mode='lines', name='Vortex Negativo', line=dict(color='red')))

        fig.update_layout(title='Indicador Vortex (VTX)',
                          xaxis_title='Fecha',
                          yaxis_title='Valor')

        fig.show()

    def ejecutar_analisis(self):
        data_df = self.cargar_datos()
        data_df = self.calcular_vortex(data_df)

        mensaje = self.interpretar_resultados(data_df)
        print("Interpretación del indicador Vortex:", mensaje)

        self.generar_grafico(data_df)

if __name__ == "__main__":
    simbolo = 'BTC-USD'
    fecha_inicio = '2024-02-05'

    analisis_vortex = AnalisisVortex(simbolo, fecha_inicio)
    analisis_vortex.ejecutar_analisis()
```
